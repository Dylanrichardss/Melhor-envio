```python
import requests
from bs4 import BeautifulSoup
import pyautogui as p
from time import sleep
# -*- coding: utf-8 -*-
```

```python
arquivo = r'C:\Users\dylan\Downloads\Melhor Envio2.html'

with open(arquivo, 'r', encoding='utf-8') as file:
    html_content = file.read()

soup = BeautifulSoup(html_content, 'html.parser')
li_items = soup.find_all('li', class_='tableContent__item')
```

```python
destinatario = []
codigo = []
sobrenome = []
sobrenomes = []
transportadora = []

for li in li_items:
    parts = li.text.split()
    b = 0
    for parte in parts:
        if parte.startswith('Destin'):
            if parte == 'Destinat√°rio':
                pass
            else:
                destinatario.append(parte[12:])
        
        elif parte.startswith('Transportadora'):
            transportadora.append(parte[14:])

        elif parte.startswith('Rastreio'):
            sobrenome = [parts[b-1],parts[b-2],parts[b-3],parts[b-4],parts[b-5],parts[b-6],parts[b-7],parts[b-8]]
            codigo.append(parts[b+1])
            v = 0

            for val in sobrenome:
                if val.startswith('Desti'):
                    sobrenome = sobrenome[:v]
                    sobrenome.reverse()
                    sobre_nome = ' '.join(sobrenome)
                    sobrenomes.append(sobre_nome)
                else:
                    pass

                v += 1
        b += 1
```

```python
b = 0
nova_lista_destinatario = []

for nome in destinatario:
    c = 0
    novo_nome = ""
    for a in nome:
        if a in ['√°', '√†', '√£', '√¢']:
            novo_nome += 'a'
        elif a in ['√©', '√®', '√™']:
            novo_nome += 'e'
        elif a in ['√≠', '√¨']:
            novo_nome += 'i'    
        elif a in ['√≥', '√¥', '√µ']:
            novo_nome += 'o'
        elif a in ['√∫', '√ª','√º']:
            novo_nome += 'u'
        elif a in ['√ß']:
            novo_nome += 'c'
        else:
            novo_nome += a
        c += 1

    nova_lista_destinatario.append(novo_nome)
    b += 1

# Agora, nova_lista_destinatario cont√©m os nomes modificados sem alterar a lista original.
```

```python
b = 0
nova_lista_sobrenomes = []

for nome in sobrenomes:
    c = 0
    novo_nome = ""
    for a in nome:
        if a in ['√°', '√†', '√£', '√¢']:
            novo_nome += 'a'
        elif a in ['√Å']:
            novo_nome += 'A'
        elif a in ['√©', '√®', '√™']:
            novo_nome += 'e'
        elif a in ['√â']:
            novo_nome += 'E'
        elif a in ['√≠', '√¨']:
            novo_nome += 'i'
        elif a in ['√ç']:
            novo_nome += 'I'       
        elif a in ['√≥', '√¥', '√µ']:
            novo_nome += 'o'
        elif a in ['√ì']:
            novo_nome += 'O'
        elif a in ['√∫', '√ª','√º']:
            novo_nome += 'u'
        elif a in ['√ö']:
            novo_nome += 'U'
        elif a in ['√ß']:
            novo_nome += 'c'
        else:
            novo_nome += a
        c += 1

    nova_lista_sobrenomes.append(novo_nome)
    b += 1

# Agora, nova_lista_destinatario cont√©m os nomes modificados sem alterar a lista original.
```

```python
destinatario = nova_lista_destinatario
sobrenomes = nova_lista_sobrenomes
```
